```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Parameters for fetching data
from_dt='29-04-2018'
to_dt='18-05-2018'
scrip_type='indexName'
scrip='NIFTY%2050'

#Create url for fetching the data from NSE website
base_url= 'https://www.nseindia.com/products/dynaContent/equities/indices/historical_pepb.jsp?'
url = base_url+scrip_type+'='+scrip+'&fromDate='+from_dt+'&toDate='+to_dt+'&yield4=all'
#print(url)

#fetch HTML from NSE website & parse the HTML
req = urllib.request.Request(url,headers={'User-Agent': 'Mozilla/5.0'})
html = urllib.request.urlopen(req).read()
parsed_html = BeautifulSoup(html,'lxml')

#Extract the needed part from parsed HTML
raw_data = parsed_html.body.find('div', attrs={'id':'csvContentDiv'}).text

#Covert data from HTML to array
l=[]
for i in raw_data.split(sep=":"):
    k=[]
    for j in i.split(sep=","):
        k.append(j.replace('"','').strip())
    l.append(k)

#Load data into Pandas DataFrame
data = pd.DataFrame(l[1:-1],columns=l[0])
for i in data.columns[1:]:
    j=i.replace('/','').replace(' ','')
    data[j]=data[i]
    del data[i]
    
#Set appropriate datatypes on the Pandas columns

data.Date = pd.to_datetime(data.Date, format="%d-%b-%Y")
data.PE = pd.to_numeric(data.PE)
data.PB = pd.to_numeric(data.PB)
data.DivYield = pd.to_numeric(data.DivYield)

data
```
